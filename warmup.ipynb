```python
import pandas as pd
import os
from env import username, host, password 
```

```python
# Acquiring telco_churn data
def get_connection(db, username=username, host=host, password=password):
    '''
    Creates a connection URL
    '''
    return f'mysql+pymysql://{username}:{password}@{host}/{db}'

## new tsa

def new_tsa():
    '''
    Returns tsa item into a dataframe
    '''
    sql_query = '''select * from stores
    join sales using(store_id)
    leftjoin items using(item_id)'''
    df = pd.read_sql(sql_query, get_connection('tsa_item_demand'))
    return df 
```

```python
def get_new_tsa():
    '''get connection, returns tsa into a dataframe and creates a csv for us'''
    if os.path.isfile('warmup.csv'):
        df = pd.read_csv('warmup.csv', index_col=0)
    else:
        df = new_tsa()
        df.to_csv('warmup.csv')
    return df
```
